```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
x_scaled = scaler.fit_transform(df[columns])
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(n_neighbors=5, metric='cosine')
knn.fit(x_scaled)
```

```python
target_calories = 500   # e.g., 500 kcal per meal
target_fat = 10         # e.g., 10 grams of fat
target_carbs = 50       # e.g., 50 grams of carbohydrates
target_protein = 25     # e.g., 25 grams of protein

# For the nutritional features not directly targeted by the user,
# use the average values from the dataset.
target_saturated_fat = df["SaturatedFatContent"].mean()
target_cholesterol   = df["CholesterolContent"].mean()
target_sodium        = df["SodiumContent"].mean()
target_fiber         = df["FiberContent"].mean()
target_sugar         = df["SugarContent"].mean()
```

```python
user_target = [500, 10, 3, 50, 400, 50, 10, 10, 35]
target_scaled = scaler.transform([user_target])
```

```python
distances, indices = knn.kneighbors(target_scaled)
```
